```python
# Import matplotlib and feature pyplot

from matplotlib import pyplot as plt
```

### Example: Line Graph

```python
x = [1, 2, 3]
y = [1, 4, 9]
plt.plot(x, y)
plt.title('Test Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Example: Two Lines on Graph

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title('Test Plot')
plt.xlabel('x')
plt.ylabel('y and z')
plt.show()
```

**Add a legend**

```python
x = [1, 2, 3]
y = [1, 4, 9]
plt.plot(x, y)
plt.title('Test Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(["this is x"])
plt.show()
```

## Compare the Population Growth in the US and China

```python
us = country[country.country == 'United States']
```

```python
us
```

```python
china = country[country.country == 'China']
```

```python
china
```

```python
plt.plot(us.year, us.population / 10**6) # divide by 1 million
plt.plot(china.year, china.population / 10**6)
plt.legend(['US', 'China'])
plt.xlabel('Year')
plt.ylabel('Population by Millions')
plt.show()
```

### What is the percentage growth of population in the US and China?

**One method is set the first year to 100 as a baseline**

```python
# Divide the US Population by the first year Population
us.population
```

```python
us.population.iloc[0]
```

```python
us.population / us.population.iloc[0] * 100
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100) # divide by 1 million
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.legend(['US', 'China'])
plt.xlabel('Year')
plt.ylabel('Population Growth (first year = 100)')
plt.show()
```
